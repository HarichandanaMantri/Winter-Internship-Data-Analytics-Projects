```python
# Question 4: Salary Formatting
# Assuming Salary column needs formatting, e.g., removing '$' and converting to numeric
df['Salary'].value_counts()
```

```python
df['Salary'] = df['Salary'].str.replace('$', '').str.replace('k', '000')
```

```python
df['Salary'] = df['Salary'].str.replace('-', ' ').str.split(' ').apply(lambda x: (float(x[0]) + float(x[1])) / 2)
```
